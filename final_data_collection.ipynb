```python
#Request data for each artist's top 10 tracks
top_tracks_data = {
    'track_id':[],
    'artist_id':[],
    'track_name':[],
    'track_release_date':[],
    'track_popularity':[]
}

for id in artist_data['artist_id']:
    pop_tracks=sp.artist_top_tracks(id)
    for i in range(len(pop_tracks['tracks'])):
        top_tracks_data['track_id'].append(pop_tracks['tracks'][i]['id'])
        print(pop_tracks['tracks'][i]['id'])
        top_tracks_data['artist_id'].append(pop_tracks['tracks'][i]['artists'][0]['id'])
        print(pop_tracks['tracks'][i]['artists'][0]['id'])
        top_tracks_data['track_name'].append(pop_tracks['tracks'][i]['name'])
        print(pop_tracks['tracks'][i]['name'])
        top_tracks_data['track_release_date'].append(pop_tracks['tracks'][i]['album']['release_date'])
        print(pop_tracks['tracks'][i]['album']['release_date'])
        top_tracks_data['track_popularity'].append(pop_tracks['tracks'][i]['popularity'])
        print(pop_tracks['tracks'][i]['popularity'])
        print('---------------------------------------------')
```

```python
#Some of our artists have top tracks in which they just feature on, meaning these tracks belong to other artists and 
# are designated with artist ids not present in our artist table. To retain data integrity, rows containing these invalid artist
# ids must be deleted.

# Create copy of tracks dataframe to iterate through
dummy_top_tracks_data_df = pd.DataFrame(top_tracks_data)

#For loop removing rows with invalid artits ids from our orignal tracks dataframe
for i in range(len(dummy_top_tracks_data_df.iloc[:,1])):
    id = dummy_top_tracks_data_df.iloc[i,1]
    exists = False
    for j in range(len(artist_data['artist_id'])):
        if id == artist_data['artist_id'][j]:
            exists = True
    if exists == False:
        top_tracks_data_df = top_tracks_data_df.drop(dummy_top_tracks_data_df.index[i])
top_tracks_data_df   
```
